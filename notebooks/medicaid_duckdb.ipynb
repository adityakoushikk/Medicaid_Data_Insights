```python
import duckdb
con = duckdb.connect()
```

```python
# Median months active per provider (count of distinct CLAIM_FROM_MONTH)
con.sql("""
SELECT MEDIAN(months_active) AS median_months_active
FROM (
    SELECT BILLING_PROVIDER_NPI_NUM, COUNT(DISTINCT CLAIM_FROM_MONTH) AS months_active
    FROM '../data/datasets/medicaid-provider-spending.csv'
    GROUP BY BILLING_PROVIDER_NPI_NUM
)
""").show()
```

```python
# Median first billing date (earliest CLAIM_FROM_MONTH) across all providers
con.sql("""
SELECT MEDIAN(first_billing::DATE) AS median_start_billing_date
FROM (
    SELECT BILLING_PROVIDER_NPI_NUM, MIN(CLAIM_FROM_MONTH || '-01') AS first_billing
    FROM '../data/datasets/medicaid-provider-spending.csv'
    GROUP BY BILLING_PROVIDER_NPI_NUM
)
""").show()
```
